```python
# STEP 1: Import Required Libraries
!pip install pymannkendall

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pymannkendall as mk
import warnings

warnings.filterwarnings("ignore", message="Could not infer format*")
sns.set(style="whitegrid")

# STEP 2: Load the CSV File
was_df = pd.read_csv("WASInput.csv")

# STEP 3: Clean Column Names
was_df.columns = was_df.columns.str.strip()
was_df.columns = was_df.columns.str.replace(r'\s*\(.*?\)', '', regex=True)
was_df.columns = was_df.columns.str.replace('.', '', regex=False)
was_df.columns = was_df.columns.str.replace('#', 'number')
was_df.columns = was_df.columns.str.replace('/', '_')
was_df.columns = was_df.columns.str.replace(' ', '_')

# STEP 4: Preview and Explore the Data
print("Shape of dataset:", was_df.shape)
print("\nColumns:\n", was_df.columns.tolist())
print("\nInfo:")
was_df.info()
print("\nSummary Stats:")
display(was_df.describe())

# STEP 5: Handle Missing and Duplicate Data
print("\nMissing Values:\n", was_df.isnull().sum()[was_df.isnull().sum() > 0])
print("\nDuplicate Rows:", was_df.duplicated().sum())

was_df = was_df.loc[:, ~was_df.columns.str.contains('^Unnamed')]
was_df = was_df.drop_duplicates()
threshold = 0.95
was_df = was_df.loc[:, was_df.isnull().mean() < threshold]

numeric_cols = was_df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    was_df[col] = was_df[col].fillna(was_df[col].median())

print("\nPost-cleaning shape:", was_df.shape)
print("Remaining missing values:", was_df.isnull().sum().sum())

# STEP 6: Correlation Heatmap for a Selected Element
selected_element = 'CO'  # â† Change this to your desired variable

if selected_element not in was_df.columns:
    print(f"\nColumn '{selected_element}' not found in cleaned columns.")
else:
    # Compute correlation with all numeric columns
    correlation_series = was_df[numeric_cols].corr()[selected_element].drop(selected_element)
    
    # Save to DataFrame
    correlation_df = correlation_series.reset_index()
    correlation_df.columns = ['Variable', f'Correlation_with_{selected_element}']

    # Print correlation values
    print(f"\nCorrelation of '{selected_element}' with other variables:")
    print(correlation_df)

    # Heatmap
    plt.figure(figsize=(10, 1 + 0.5 * len(correlation_series)))
    sns.heatmap(correlation_series.to_frame().T, annot=True, cmap='coolwarm', center=0, cbar=True, fmt=".2f")
    plt.title(f"Correlation of '{selected_element}' with Other Variables")
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
